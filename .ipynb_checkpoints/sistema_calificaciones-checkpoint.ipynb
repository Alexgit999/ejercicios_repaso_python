# Ejercicio: Sistema de Calificaciones
Descripción del problema:
Vas a crear un programa que permita a un profesor gestionar las calificaciones de sus alumnos.
## El programa debe:
Pedir al usuario que introduzca el número de estudiantes.
Solicitar el nombre y la calificación de cada estudiante (la calificación debe ser un número entre 0 y 100).
Guardar esta información en una estructura de datos adecuada.
Mostrar un menú que permita:
1. Ver la lista completa de estudiantes y sus calificaciones.
2. Mostrar la calificación promedio del grupo.
3. Mostrar al estudiante con la calificación más alta y al estudiante con la calificación más baja.
4. Permitir buscar la calificación de un estudiante por su nombre.
5. Salir del programa.
# Requisitos:
1. Validación de entrada:
La calificación debe estar entre 0 y 100. Si no es válida, solicita al usuario que la ingrese de nuevo.
El nombre del estudiante debe ser un texto no vacío.

4. Menú interactivo:

```python
## import sys

# Calcular promedio:
def promedio(punt):
    print("##### El promedio de notas de la clase es: ", sum(punt)/len(punt))

# Encontrar puntuación más alta y más baja:
def max_min(nombre, punt, nt):
    maxi = 0
    mini = 0
    for i in range(nt):
        if punt[i] > punt[maxi]:
            maxi = i
        if punt[i] < punt[mini]:
            mini = i
    print("##### El estudiante con la calificacion más alta es: ", nombre[maxi])
    print("##### El estudiante con la calificacion más baja es: ", nombre[mini])
    #menu(nombre, punt, nt, breaker)

# Encontrar un estudiante y su nota por nombre
def buscar (nombre, punt, nt):
    imputado = str(input("¿Cual es el nombre del estudiante que desea encontrar?"))
    i = 0
    while i < nt and i + 1 != None:
        if nombre[i] == imputado:
            print("##### El estudiante ", imputado, " tiene una calificación de ", punt[i])
        return
    print("##### El nombre solicitado no existe en la lista")

# Funcion de menu
def menu(nombre, punt, nt, breaker):
    while True:
        me = int(input("\nQué desea hacer ahora?\n\n( 1 ) Mostrar la calificación promedio del grupo.\n( 2 ) Mostrar al estudiante con la calificación más alta y al estudiante con la calificación más baja.\n( 3 ) Permitir buscar la calificación de un estudiante por su nombre.\n( 4 ) Salir del programa.\n\n"))
        if me == 1:
            promedio(punt)
            #menu(nombre, punt, nt, breaker)
        elif me == 2:
            max_min(nombre, punt, nt)
            #menu(nombre, punt, nt, breaker)
        elif me == 3:
            buscar(nombre, punt, nt)
            #menu(nombre, punt, nt, breaker)
        elif me == 4:
            print("##### Saliendo del sistema")
            breaker = False
            break
        else:
            print("##### Elija una opción correcta")
    return


## Inicio

nt = int(input("¿Cuántos estudiantes tiene su clase?"))
nombre = [""] * nt
punt = [0] * nt
breaker = True

i = 0
while i < nt :
    nombre[i] = str(input(f"Introduzca nombre del estudiante con ID {i}\n"))
    punt[i] = int(input("introduzca su puntuación: "))
    i = i + 1

if breaker == True:
    menu(nombre, punt, nt, breaker)
else:
    exit()
```
